```python
k = np.arange(0, 101)
x = np.sin(2*np.pi*0.07*k)

plt.figure(figsize=[12, 4]) 

plt.plot(k, x, color='red', alpha=0.75, lw=1.5, ls='--')
plt.stem(k, x)

# сетка
plt.minorticks_on()
plt.grid(which = 'major', color = 'black', linewidth = 1, linestyle = '-', alpha = 0.75)
plt.grid(which = 'minor', color = 'grey', linewidth = 1, linestyle = '--', alpha = 0.5)

plt.xlabel('k')
plt.ylabel('x[k]')
plt.title('График зависимости x[k]')
```

```python
def my_fun(x: np.ndarray):
    np.seterr(invalid='ignore')
    res = np.where(x == 0, 1, np.sin(x)/x)
    return res
```

```python
y = my_fun(k)

plt.figure(figsize=[12, 4]) 

plt.plot(k, y, color='blue', alpha=0.75, lw=1.5, ls='-')

# сетка
plt.minorticks_on()
plt.grid(which = 'major', color = 'black', linewidth = 1, linestyle = '-', alpha = 0.75)
plt.grid(which = 'minor', color = 'grey', linewidth = 1, linestyle = '--', alpha = 0.5)

plt.xlabel('k')
plt.ylabel('y[k]')
plt.title('График зависимости y[k]')
```

### Различие аналогового, дискретного и цифрового сигнала

- Аналоговый сигнал -- обычный сигнал, функция времени, имеющая бесконечное количество значений (может принимать всевозможные значения) в пределах некоторого промежутка.
- Дискретный сигнал -- счетный набор отсчетов в заданные моменты времени $k \Delta t$, где $\Delta t$ заданное значение. Дискретный сигнал все еще может принимать любые значения в пределах заданного промежутка, но уже не является непрерывной функцией времени
- Цифровой сигнал. Дискретный сигнал, прошедший процедуру квантования. Значение такого сигнала уже не может принимать любый значения в пределах заданного промежутка. Более того, значение аналовогого и цифрового сигнала в некоторой точке $t = k \Delta t$ могут не совпадать (из-за того, что цифровой сигнал квантуется).

### Спектр одиночного прямоугольного импльса

$$x(t) = 1 \hspace{5mm} |t| < \tau$$

$$X(f) = \tau \frac{\sin(\pi f \tau)}{\pi f \tau}$$

```python
# зададим величину tau
tau = 1

# здадим функцию спектра X(f)
def X(f):
    return tau * np.sin(np.pi * f * tau) / (np.pi * f * tau) 

# зададим диапазон частот
f = np.linspace(-5/tau, 5/tau, 1000)
plt.plot(f, X(f), color='blue', alpha=0.75, lw=1.5, ls='--')


# сетка
plt.minorticks_on()
plt.grid(which = 'major', color = 'black', linewidth = 1, linestyle = '-', alpha = 0.75)
plt.grid(which = 'minor', color = 'grey', linewidth = 1, linestyle = '--', alpha = 0.5)

plt.xlabel('f, Hz')
plt.ylabel('X(f)')
plt.title('График зависимости X(f)')
```

###  Спектр синусоидального сигнала

Синусоида раскладывается на две мнимые экспоненты, таким образом ее спектр будет состоять из двух симметрично расположенных дельта функций.

$$x(t) = \sin(\omega t)$$

$$X(f) = -\frac{i}{2} \delta(f - \frac{\omega}{2 \pi}) + \frac{i}{2} \delta(f + \frac{\omega}{2 \pi})$$

```python
omega = 1 

plt.plot([-omega / (2 * np.pi), -omega / (2 * np.pi)], [0, 0.5])
plt.plot([omega / (2 * np.pi), omega / (2 * np.pi)], [0, -0.5])

# сетка
plt.minorticks_on()
plt.grid(which = 'major', color = 'black', linewidth = 1, linestyle = '-', alpha = 0.75)
plt.grid(which = 'minor', color = 'grey', linewidth = 1, linestyle = '--', alpha = 0.5)
```

### Спектр треуольного импульса 

$$X(f) = \frac{\tau}{2} \frac{\sin^2(\pi f \tau / 2)}{(\pi f \tau / 2)^2}$$

```python
tau = 1

def spectrum(f):
    return tau / 2 * (np.sin(np.pi * f * tau / 2)**2) / ((np.pi * f * tau / 2)**2) 

f = np.linspace(-5/tau, 5/tau, 1000)
plt.plot(f, spectrum(f))

# сетка
plt.minorticks_on()
plt.grid(which = 'major', color = 'black', linewidth = 1, linestyle = '-', alpha = 0.75)
plt.grid(which = 'minor', color = 'grey', linewidth = 1, linestyle = '--', alpha = 0.5)

plt.xlabel('f, Hz')
plt.ylabel('X(f)')
plt.title('График зависимости X(f)')
```
